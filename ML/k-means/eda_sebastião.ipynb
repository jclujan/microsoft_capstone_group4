```python
%sql
SELECT
    buyer_name,
    buyer_country,
    cluster_id,
    total_contracts,
    ROUND(single_bidder_rate * 100, 1) AS single_bidder_pct,
    ROUND(avg_award_value_eur / 1e6, 2) AS avg_contract_M_eur
FROM workspace.gold.buyer_profiles_clustered
WHERE top_cpv_division = '72'
  AND single_bidder_rate < 0.2
ORDER BY avg_award_value_eur DESC
LIMIT 20
```
